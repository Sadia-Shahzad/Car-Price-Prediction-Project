### Converting the categorical columns in the numerical columns to find correlations and Multicollinearity

```python
Fuel_Type=[0 if i=='Petrol' else 1 for i in df['Fuel_Type']]
Seller_Type=[0 if i=='Dealer' else 1 for i in df['Seller_Type']]
Transmission=[0 if i=='Manual' else 1 for i in df['Transmission']]
```
