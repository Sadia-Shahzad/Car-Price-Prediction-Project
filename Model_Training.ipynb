## Training the model

Now time for model Training

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('Car Price Prediction Dataset/Car_Dataset_Cleaned.csv')
```

```python
Y=data['Selling_Price']
```

```python
Y
```

```python
X=data.drop("Selling_Price",axis=1)
```

```python
X
```

```python
X.drop('Car_Name',axis=1,inplace=True)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
Y_test.shape
```

```python
Y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

```python
plt.subplots(figsize=(22, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

## Model Training

### Multiple Linear Regression Problem

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

model=LinearRegression()
model.fit(X_train_scaled,Y_train)
y_test_pred=model.predict(X_test_scaled)

mae=mean_absolute_error(Y_test,y_test_pred)
test_r2=r2_score(Y_test,y_test_pred)

print("mean absolute error is ",mae)
print("r2_score is ",test_r2)


y_train_pred=model.predict(X_train_scaled)

plt.scatter(Y_train,y_train_pred)
```

```python
#checking if the model is overfitting
y_train_pred=model.predict(X_train_scaled)
train_r2=r2_score(Y_train,y_train_pred)
print("train r2 score is ",train_r2," whereas test r2 score is ",test_r2)
```

The gap is 0.15 or 15% so definitely model perfroms well with train data and less well with test data . though it would be always like trainr2 > test r2 but the amount of difference matters . it is not disastrous overfitting but not too less either.So will fix it using some regularization techniques (i.e L2 regularization or Ridge regression algorithm)

### Hyperparameter tuning the model using Ridge Regression Algorithm with Cross Validation to avoid overfitting

# Ridge Regression Algorithm with Cross-Valiation

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridge=Ridge()
ridge.fit(X_train_scaled,Y_train)
ridge_y_pred=ridge.predict(X_test_scaled)
mae=mean_absolute_error(Y_test,ridge_y_pred)
ridge_r2=r2_score(Y_test,ridge_y_pred)

print("MAE=",mae)
print("R^2 score=",ridge_r2)
```

### RidgeCV

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
ridgeCV=RidgeCV()
ridgeCV.fit(X_train_scaled,Y_train)
ridge_y_pred=ridgeCV.predict(X_test_scaled)
mae=mean_absolute_error(Y_test,ridge_y_pred)
ridge_r2=r2_score(Y_test,ridge_y_pred)

print("MAE=",mae)
print("R^2 score=",ridge_r2)
```

```python
#checking the Overfitting
ridge_ytrain_pred=ridgeCV.predict(X_train_scaled)

train_ridge_r2=r2_score(Y_train,ridge_ytrain_pred)

print("train r2 for ridge hyperparameterized tuned LR model using CV is",train_ridge_r2,"test r2 is",ridge_r2)
```

75% accuracy is achieved
